```python
import cv2
import mediapipe as mp

# Check OpenCV CUDA support
count = cv2.cuda.getCudaEnabledDeviceCount()
print("OpenCV CUDA support:", count > 0, f"(Devices: {count})")

# Check MediaPipe (MediaPipe generally uses CPU in standard Python pip packages,
# or OpenGL/Metal for GPU, rather than native CUDA directly, but we can verify it imports successfully)
print("MediaPipe version:", mp.__version__)
try:
    from mediapipe.tasks import python
    print("MediaPipe GPU Delegate option available:", hasattr(python.BaseOptions.Delegate, "GPU"))
except Exception as e:
    print("MediaPipe GPU Delegate check failed:", e)
```
